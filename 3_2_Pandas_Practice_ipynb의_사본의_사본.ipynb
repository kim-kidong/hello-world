#Q16. DataFrame 병합
---
`df0110`은 2001-2010년, `df1120`은 2011-2020년 용도별 전력 판매량 입니다.  
두 DataFrame을 병합하여 `df0120`이라는 이름으로 저장하고 출력하세요.  
*Hint. pd.concat()*

```python
# 문제의 입력값입니다. 수정하지 마세요.
url = "https://docs.google.com/uc?export=download&id="
df0110 = pd.read_csv(
    url + "1_lhuWUoEk05YMtw9LuuLjOOH_ahVBsfh", encoding="cp949"
)
df1120 = pd.read_csv(
    url + "1j6X33UDx1of6Eyp8ZaH86A-aACtmV_9H", encoding="cp949"
)
```

```python
# 아래에 정답코드를 작성해주세요.
df0120 = pd.concat([df0110 , df1120]).sort_values("연도")

df0120.reset_index(drop=True)
```

#Q17. DataFrame join
---
*air_quality_pm25* 는 미세먼지, *air_quality_no2*는 이산화질소 측정 값입니다.  
*air_quality_pm25*의 "date.utc", "location" 컬럼 기준으로 두 DataFrame을 join 하고 `air_quality`라는 이름으로 저장하세요.  
*Hint. pd.merge() how=[left || right], on="컬럼"*

```python
# 문제의 입력값입니다. 수정하지 마세요.
url = "https://docs.google.com/uc?export=download&id="
air_quality_pm25 = pd.read_csv(
    url + "1O2o5h277w6MoBYjbgOHCPtfbhABZ8XRB",
    parse_dates=True,
    usecols=["date.utc", "location", "parameter", "value"]
)

air_quality_no2 = pd.read_csv(
    url + "1nc4toKu_5ywrcD71Log_EYlZP6-6tby7",
    parse_dates=True,
    usecols=["date.utc", "location", "parameter", "value"]
)
```

```python
# 아래에 정답코드를 작성해주세요.

pd.merge(air_quality_pm25, air_quality_no2, how="right", on = ["date.utc", "location" ])
```
